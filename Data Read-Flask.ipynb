```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import urllib
import matplotlib.dates as mdates
```

```python
# File to Load (Remember to Change These)
file_to_load = "Resources/Zillow San Diego Value - Sheet1.csv"

# Read Purchasing File and store into Pandas data frame
zillow_price_df = pd.read_csv(file_to_load)
```

```python
zillow_price_df.head(203)
```

```python
zillow_price_df=zillow_price_df.dropna()
zillow_price_df
```

```python
zillow_price_df["CityState"]= zillow_price_df["RegionName"]+ ", " + zillow_price_df["State"]

zillow_price_df.head()
```

```python
zillow_price_df=zillow_price_df.drop(columns=["RegionID","RegionName","State", "Metro","CountyName", "2003-10","2003-11","2003-12","2020-01","2020-02"])
```

```python
zillow_price_df= zillow_price_df.set_index("CityState")
zillow_price_df.head(41)
```

```python
for i in range(2004,2020):
    zillow_price_df[i] = round((zillow_price_df[str(i)+"-01"]+
                                zillow_price_df[str(i)+"-02"]+
                                zillow_price_df[str(i)+"-03"]+
                                zillow_price_df[str(i)+"-04"]+
                                zillow_price_df[str(i)+"-05"]+
                                zillow_price_df[str(i)+"-06"]+
                                zillow_price_df[str(i)+"-07"]+
                                zillow_price_df[str(i)+"-08"]+
                                zillow_price_df[str(i)+"-09"]+
                                zillow_price_df[str(i)+"-10"]+
                                zillow_price_df[str(i)+"-11"]+
                                zillow_price_df[str(i)+"-12"])/12,2)
    zillow_price_df = zillow_price_df.drop(columns = [str(i)+"-01",
                                   str(i)+"-02",
                                   str(i)+"-03",
                                   str(i)+"-04",
                                   str(i)+"-05",
                                   str(i)+"-06",
                                   str(i)+"-07",
                                   str(i)+"-08",
                                   str(i)+"-09",
                                   str(i)+"-10",
                                   str(i)+"-11",
                                   str(i)+"-12"])
zillow_price_df.head()
```

```python
zillow_price_dft= zillow_price_df.T

zillow_price_dft.head(17)
```
